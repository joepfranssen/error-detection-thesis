```python
import os
files = os.listdir('.')
```

```python
import pandas as pd
import fitz  # this is pymupdf
df_non_retracted = pd.DataFrame(columns=['DOI', 'Content', 'Retracted']) #make empty dataframe
for study in files: #iterate through the files
  try:
    with fitz.open(study) as doc: #open the document
        text = "" #start with empty string
        for page in doc: #iterate through the pages of the file
            text += page.get_text() #add the found text to the empty string
        df_non_retracted = df_non_retracted.append({'DOI': study, 'Content': text, 'Retracted': 0}, ignore_index=True) #add DOI, text and 0 for non-retracted paper to the dataframe
  except: #if anything above fails, pass this file
    pass
```
